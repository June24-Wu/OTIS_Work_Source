```python
import os, uuid
import time, datetime
import multiprocessing
from multiprocessing import cpu_count
import pandas as pd

from azure.cosmos import CosmosClient
```

```python
container_name=""
storage_account_name = ""
storage_account_access_key = ""

connect_str = ""

file_location = f"wasbs://{container_name}@{storage_account_name}.blob.core.windows.net/"
spark.conf.set(
  "fs.azure.account.key."+storage_account_name+".blob.core.windows.net",
  storage_account_access_key)
```

```python
cameras = dbutils.fs.ls(file_location)
data = [] # camera_name , year , month , day , count 
for camera in cameras:
  camera_name = camera[1]
  camera_location = file_location + camera_name  # wasbs://espdata@gw2data2lake2storage.blob.core.windows.net/00000000000000T/
  camera_name = camera_name.replace('/','') # 00000000000000T
  years = dbutils.fs.ls(camera_location)
  
  
  for year in years:
    year_name = year[1]
    year_location = camera_location + year_name # wasbs://espdata@gw2data2lake2storage.blob.core.windows.net/00000000000000T/2021/
    year_name = year_name.replace('/','')  # 2021 
    months = dbutils.fs.ls(year_location)
    
    
    for month in months:
      month_name = month[1]
      month_location = year_location + month_name # wasbs://espdata@gw2data2lake2storage.blob.core.windows.net/00000000000000T/2021/1/
      month_name = month_name.replace('/','') # 1
      days = dbutils.fs.ls(month_location)
      
      
      for day in days:
        day_name = day[1]
        day_location = month_location + day_name # wasbs://espdata@gw2data2lake2storage.blob.core.windows.net/00000000000000T/2021/1/15/
        day_name = day_name.replace('/','') # 15
        images = dbutils.fs.ls(day_location)
        
        
        # append
        row = []
        row.extend([camera_name,year_name,month_name,day_name,len(images)])
        data.append(row)
        print(row)
```

```python
data_final = pd.DataFrame(data,columns = ['camera_name' , 'year' , 'month' , 'day' , 'up_load_image_num'])
```
