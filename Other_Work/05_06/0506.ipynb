```python
import tensorflow as tf
from tensorflow.keras import layers
from PIL import Image
import os
import pathlib
import pathlib
import shutil
import numpy as np
import pandas as pd
```

```python
# DATA_PATH = "./Pictures"
img_height = 120
img_width = 160
seed = 12

# batch_size = 128
# resized_height = 120
# resized_width = 160
num_classes = 2
NUM_EPOCHS = 120
# STEPS_PER_EPOCH = 1508 //batch_size
```

```python
model = tf.keras.models.load_model('./Best_Model')
model.summary()
```

```python
path = './Pictures'
files_name = os.listdir(path)
img_list= []
test = []
for files in files_name:
    img_path = path + "/" + files
    img_list.append(files)
    print(img_path)
    image = tf.keras.preprocessing.image.load_img(img_path,
                                                  grayscale=False,
                                                  color_mode='rgb',
                                                  target_size=(img_height, img_width),
                                                  interpolation='bilinear'
                                                 )
    input_arr = tf.keras.preprocessing.image.img_to_array(image)
    input_arr = np.array([input_arr])
    input_arr = (input_arr/127.5) - 1
    test.append(input_arr)
test = np.array(test)
test = test.reshape(test.shape[0],120,160,3)
```

```python
result = model.predict(test)
result
```

```python
predict_list = []
for row in result:
    l = []
    score = int(row[1]*255)
    l.append(score)
    if score > 128:
        l.append(1)
    else:
        l.append(0)
    predict_list.append(l)
predict_list
```

```python
total_df = []
for i in range(len(predict_list)):
    df =[]
    df.append(img_list[i]) # img
    df.append(predict_list[i][1])  # computer answer
    df.append(predict_list[i][0])  # computer score
    total_df.append(df)
total_df = pd.DataFrame(total_df,columns=['img',
                                          'computer answer',
                                          'computer score',
                             ])
total_df
```

```python
import shutil
base_path = "./Computer_Classification"
def create_examination(base_path):
    os.makedirs(base_path)
    os.makedirs(base_path+"/positive")
    os.makedirs(base_path+"/negative")
    return None

if not os.path.exists(base_path):
    create_examination(base_path)
else:
    shutil.rmtree(base_path)
    create_examination(base_path)

positive = total_df[total_df['computer answer']==1]
negative = total_df[total_df['computer answer']==0]
for i in positive['img']:
    shutil.copy(path + "/" + i, base_path+"/positive/"+i)
    positive.to_csv(base_path+"/positive/Positive Result.csv")
for i in negative['img']:
    shutil.copy(path + "/" + i, base_path+"/negative/"+i)
    negative.to_csv(base_path+"/negative/Negtive Result.csv")
```
