```python
import tensorflow as tf
from tensorflow.keras import layers
from PIL import Image
import os
import pathlib
import pathlib
import shutil
import numpy as np
import pandas as pd
```

```python
import shutil
base_path = "./Computer_Classification"
def create_examination(base_path):
    os.makedirs(base_path)
    os.makedirs(base_path+"/positive")
    os.makedirs(base_path+"/negative")
    return None

if not os.path.exists(base_path):
    create_examination(base_path)
else:
    shutil.rmtree(base_path)
    create_examination(base_path)
```

```python
path = './Pictures'
files_name = os.listdir(path)
count = 0 
for files in files_name:
    img_path = path + "/" + files
    score = int(files.split('--')[-1].replace('.jpg',''))
    if score > 128:
        shutil.copy(img_path, base_path+"/positive/"+ files)
    else:
        shutil.copy(img_path, base_path+"/negative/"+files)
    count += 1
    print(count)
    print(img_path)
```
