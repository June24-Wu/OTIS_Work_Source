```python
DATA_PATH = "./data"
img_height = 120
img_width = 160
seed = 12

# batch_size = 128
# resized_height = 120
# resized_width = 160
num_classes = 2
NUM_EPOCHS = 120
# STEPS_PER_EPOCH = 1508 //batch_size
```

```python
def get_model(lr=0.001, alpha=0.35):
    base_model = tf.keras.applications.MobileNetV2(
        input_shape=(img_height, img_width, 3),
        include_top=False, weights='imagenet', alpha=alpha)
#     base_model.trainable = False
    base_model.trainable = True
    # Add a new classifier layer for transfer learning
    global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
    dropout = tf.keras.layers.Dropout(0.2)
    prediction_layer = tf.keras.layers.Dense(num_classes, activation='softmax')
#     prediction_layer = tf.keras.layers.Dense(1, activation='softmax')
  
    model = tf.keras.Sequential([
      base_model,
      global_average_layer,
      dropout,
      prediction_layer
    ])
    return model
  
def get_compiled_model(alpha=0.35, lr=0.001):
    model = get_model(lr, alpha)
#     lr = learning_rate = CustomSchedule(2)
    model.compile(
                optimizer=tf.keras.optimizers.SGD(learning_rate=lr, momentum=0.9),
#                 optimizer=tf.keras.optimizers.SGD(learning_rate=0.01),
                loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                metrics=['accuracy'])
    return model
```

```python
model = get_compiled_model()
checkpoint_path = "./model/cp-0066.ckpt"
model.load_weights(checkpoint_path)
```

```python
path = 'D:/PyCharmproject/OTIS/03_23/test/data'
files_name = os.listdir(path)[1:]
files_name
img_list = []
img_score_list =[]
test = []
result_list2 =[]
for files in files_name:
    sub_path = path + "/" + files
    sub_files_name = os.listdir(sub_path)
    for img in sub_files_name:
        if files == 'positive':
            result_list2.append(1)
        else:
            result_list2.append(0)
        img_path = sub_path + "/" + img
        img_list.append(img_path)
        print(img_path)
        image = tf.keras.preprocessing.image.load_img(img_path,
                                                      grayscale=False,
                                                      color_mode='rgb',
                                                      target_size=(img_height, img_width),
                                                      interpolation='bilinear'
                                                     )
        input_arr = tf.keras.preprocessing.image.img_to_array(image)
        input_arr = np.array([input_arr])
        input_arr = (input_arr/127.5) - 1
        test.append(input_arr)
        
        img_score = img.split("-")
        img_score_list.append(int(img_score[-1][:3]))
test = np.array(test)
test = test.reshape(test.shape[0],120,160,3)
```

```python
result = model.predict(test)
result
```

```python
predict_list = []
for row in result:
    l = []
    score = int(row[1]*255)
    l.append(score)
    if score > 128:
        l.append(1)
    else:
        l.append(0)
    predict_list.append(l)
predict_list
```

```python
result_list = []
for i in img_score_list:
    l = []
    l.append(i)
    if i > 128:
        l.append(1)
    else:
        l.append(0)
    result_list.append(l)
result_list
```

```python
total_df = []
for i in range(len(result_list)):
    df =[]
    df.append(img_list[i].replace('D:/PyCharmproject/OTIS/03_23/test/data/','')) # img
    df.append(result_list2[i])  # True answer
    df.append(predict_list[i][1])  # predict answer
    df.append(result_list[i][1])  # camera answer
    df.append(predict_list[i][0])  # predict score
    df.append(result_list[i][0])  # camera score
    total_df.append(df)
total_df = pd.DataFrame(total_df,columns=['img','true answer','computer answer',
                              'camera answer','computer score',
                              'camera score'
                             ])
total_df
```

```python
def split_by_score(path, dest, score_thr):
    images = os.listdir(path)
    for img in images:
        score = float(os.path.splitext(img)[0].split('-')[-1])
        if score >= score_thr:
            shutil.copy(path+'/'+img, dest+'/positive/'+img)
        else:
            shutil.copy(path+'/'+img, dest+'/negative/'+img)
```
