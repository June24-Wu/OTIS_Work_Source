```python
import shutil
base_path = "./Classification"
def create_examination(base_path):
    os.makedirs(base_path)
    os.makedirs(base_path+"/P")
    os.makedirs(base_path+"/N")
    return None

if not os.path.exists(base_path):
    create_examination(base_path)
else:
    shutil.rmtree(base_path)
    create_examination(base_path)
```

```python
path = './Picture'
files_name = os.listdir(path)
for score in files_name:
    files = path + '/' + score
    scores = int((score.split('--')[-1]).replace('.jpg',''))
    print(scores)
    if scores >= 128:
        shutil.copy( files, base_path+"/P/"+score)
    else:
        print(files)
        shutil.copy( files, base_path+"/N/"+score)
```
