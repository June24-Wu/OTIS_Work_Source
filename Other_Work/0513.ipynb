```python
import tensorflow as tf
from tensorflow.keras import layers
from PIL import Image
import os
import pathlib
import pathlib
import shutil
import numpy as np
import pandas as pd
```

```python
# DATA_PATH = "./Pictures"
img_height = 120
img_width = 160
seed = 12

# batch_size = 128
# resized_height = 120
# resized_width = 160
num_classes = 2
NUM_EPOCHS = 120
# STEPS_PER_EPOCH = 1508 //batch_size
```

```python
model = tf.keras.models.load_model('./Best_Model')
model.load_weights('./Model/Best_Model.ckpt')
model.summary()
```

```python
path = './Pictures'
files_name = os.listdir(path)
final = []
test = []
for files in files_name:
    img_path = path + "/" + files
    pic_list = os.listdir(img_path)
    for pic in pic_list:
        final_list = [] # pic_path ,pic_name, real , camera_score
        pic_path = img_path + "/" + pic
        final_list.append(pic_path)
        final_list.append(pic)
        if files == 'positive':
            final_list.append(1)
        else:
            final_list.append(0)
        ai_camera_score = int(pic.split('--')[-1].replace('.jpg',''))
        
        if ai_camera_score >= 128:
            final_list.append(1)
        else:
            final_list.append(0)
        final.append(final_list)

        image = tf.keras.preprocessing.image.load_img(pic_path,
                                                      grayscale=False,
                                                      color_mode='rgb',
                                                      target_size=(img_height, img_width),
                                                      interpolation='bilinear'
                                                     )
        input_arr = tf.keras.preprocessing.image.img_to_array(image)
        input_arr = np.array([input_arr])
        input_arr = (input_arr/127.5) - 1
        test.append(input_arr)
test = np.array(test)
test = test.reshape(test.shape[0],120,160,3)
final = pd.DataFrame(final)
test
```

```python
result = model.predict(test)
result
```

```python
predict_list = []
for row in result:
    l = []
    score = int(row[1]*255)
    l.append(score)
    if score > 128:
        l.append(1)
    else:
        l.append(0)
    predict_list.append(l)
predict_list
```

```python
computer = pd.DataFrame(predict_list)
df = pd.concat([final,computer],axis=1)
df.columns =['pic_path','pic_name','real','camera','computer_score','computer']
df
```
